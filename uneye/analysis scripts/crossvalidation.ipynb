# U'n'Eye crossvalidations on datasets 1-3 published in Bellet et al. (2018)

```python
import numpy as np
import ueye
import os
```

# dataset1

```python
dset = 'dataset1'
datapath = os.path.join('../data',dset+'/')
ntrain = 950
sampfreq = 1000
classes = 2
min_sacc_dur = 6
min_sacc_dist = 10
weights_name ='dataset1_coder2'

# load training data, shuffle and choose ntrain number of samples
# load csv files
X = np.loadtxt(datapath+dset+'_'+str(sampfreq)+'hz_X_train.csv',delimiter=',')
Y = np.loadtxt(datapath+dset+'_'+str(sampfreq)+'hz_Y_train.csv',delimiter=',')
L = np.loadtxt(datapath+dset+'_'+str(sampfreq)+'hz_Labels2_train.csv',delimiter=',')
print(X.shape)

np.random.seed(1)
randind = np.random.permutation(X.shape[0])
# training data
Xtrain = X[randind[:ntrain],:]
Ytrain = Y[randind[:ntrain],:]
Ltrain = L[randind[:ntrain],:]

# validation data
Xval = X[randind[ntrain:],:]
Yval = Y[randind[ntrain:],:]
Lval = L[randind[ntrain:],:]

# Cross Validation
model = ueye.DNN(weights_name=weights_name,
                 sampfreq=sampfreq,classes=classes,
                min_sacc_dur = min_sacc_dur,
                min_sacc_dist=min_sacc_dist)

model.crossvalidate(Xtrain,Ytrain,Ltrain,Xval,Yval,Lval)
```
