# Detecting fixations

In the previous notebooks we saw that it was easy to visually distinguish the two main gaze-related events, fixations and saccades. When the gaze location is "jittering" around a small area, it's a fixation, and when gaze makes a sudden, larger jump, it's a saccade. To detect these events automatically, we might think of doing something like this:

```
events_list = []
for each gaze_point in data:
  if (distance between current gaze_point and previous gaze_point) < threshold:
    add 'fixation' to events_list
  else:
    add 'saccade' to events_list
```

In fact, many of the traditional algorithms for detecting gaze events are not that far from this simple idea.

````python
# Task 1. Your Pseudocode of your algorithm here
# (Pseudocode by Oma-Emmi Hallman, a couple of minor changes by Alpo Ranne)
# Pseudocode for I-VT algorithm
#<
#I-VT (dataset, velocity threshold) 
#Calculate point-to-point velocities for each point in the dataset
#Label each point below velocity threshold as a fixation point, otherwise as a saccade point 
#Remove saccade points by combining fixation points into groups and discard saccade point 
#Use the centroid of the groups to translate groups into a representation 
#Return fixations
#>
# The last three lines of the pseudocode were done differently here, since the data is not visualized as "fixation blobs",
# but instead with colors for fixations and saccades, so it is good to keep the saccade points in the data as well.
````

**Task 2.** Then, edit the function in the next code block so that it takes the x- and y-coordinates of the gaze, applies your fixation detection algorithm and outputs a list where each element is either 'Fixation' or 'Saccade'. The list should have the same length as the original eye tracker dataset, 1946.

The gaze locations are in the et_df dataframe under column names 'Gaze point X', 'Gaze point 'Y. One row corresponds to one measured gaze point.

Implement your algorithm in the code block. As an example, the function shows an implementation of the simple algorithm, feel free to test it with different threshold values. Replace lines 3 - 5 with you own code.

````python
# Task 2.
# python implementation of the algorithm by Alpo Ranne

def I_VT(gaze_points, time_stamps, velocity_threshold):
  # Calculate velocities
  distances = [np.linalg.norm(p_i - p_i_1) for p_i_1, p_i in zip(gaze_points, gaze_points[1:])]
  sample_durations = [(s_i - s_i_1)/1000 for s_i_1, s_i in zip(time_stamps, time_stamps[1:])]
  velocities = [s/t for s, t in zip(distances, sample_durations)]

  # Label events
  events = ['Fixation' if velocity < velocity_threshold else 'Saccade' for velocity in velocities]
  events.append(events[-1])
  return events

events = I_VT(df_et[['Gaze point X', 'Gaze point Y']].values, df_et['Recording timestamp'].values, velocity_threshold = 1.2)
print(events[-10:])
````

````python
# Additional velocity vs. time plot
# Get these again because they were defined inside a function... In hindsight probably should have done that differently
gaze_points = df_et[['Gaze point X', 'Gaze point Y']].values
distances = [np.linalg.norm(p_i - p_i_1) for p_i_1, p_i in zip(gaze_points, gaze_points[1:])]
time_stamps = df_et['Recording timestamp'].values
sample_durations = [(s_i - s_i_1)/1000 for s_i_1, s_i in zip(time_stamps, time_stamps[1:])]
velocities = [s/t for s, t in zip(distances, sample_durations)]

# Plotting
fig = plt.figure()
ax = plt.gca()
fig.set_size_inches(20,15)
plt.plot(time_stamps[:-1], velocities)
plt.xlabel('Time (ms)', fontsize=20)
plt.ylabel('velocity [unit/ms]', fontsize=20)
plt.title('I-VT algorithm velocity vs. time', fontsize=20)
plt.xticks(ticks=[0.25e7, 0.5e7, 0.75e7, 1e7, 1.25e7, 1.5e7, 1.75e7, 2e7], labels=[2500, 5000, 7500, 10000, 12500, 15000, 17500, 20000])
plt.xlim([min(time_stamps),max(time_stamps)])
plt.show()
````

````python
# Total fixation and saccade duration Tobii I-VT
df_et2_unique = df_et.drop_duplicates(subset='Eye movement type index')
df_et2_unique_fixations = df_et2_unique[df_et2_unique['Eye movement type'] == 'Fixation']
df_et2_unique_saccades = df_et2_unique[df_et2_unique['Eye movement type'] == 'Saccade']
total_fixation_duration = sum(df_et2_unique_fixations['Gaze event duration'].values)
total_saccade_duration = sum(df_et2_unique_saccades['Gaze event duration'].values)
print('Tobii I-VT:\nFixation duration: ', total_fixation_duration, '\nSaccade duration: ', total_saccade_duration)
````

````python
# total fixation and saccade durations our I-VT
fixation_duration = 0
saccade_duration = 0
for i in range(0,len(events)-1):
  if events[i] == 'Fixation':
    fixation_duration += sample_durations[i]
  else:
    saccade_duration += sample_durations[i]

print(f'Our I-VT:\nFixation duration: {fixation_duration:.1f}\nSaccade duration: {saccade_duration:.1f}')
````
